```python
import pdfplumber
import pandas as pd

pdf_path = "/content/seleksi adm.pdf"

filtered_tables = []

with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        table = page.extract_table()

        if table:
            df = pd.DataFrame(table[1:], columns=table[0])

            # Normalize column count (handle None headers)
            num_cols = len(df.columns)

            if num_cols == 7:
                print(f"Page {i+1}: ✅ Lanjut (7 columns)")
                filtered_tables.append(df)
            else:
                print(f"Page {i+1}: ❌ Skip ah ({num_cols} columns)")

# Merge all valid tables
if filtered_tables:
    df1 = pd.concat(filtered_tables, ignore_index=True)
else:
    df1 = pd.DataFrame()

print("Final shape:", df1.shape)

df1.to_excel("df1_original_file.xlsx", index=False)
```

```python
import pandas as pd
from rapidfuzz import process, fuzz

# Install dulu jika belum: pip install rapidfuzz

def fuzzy_merge(df2, df1, col='Nama', threshold=0):
    """
    Merge df2 dengan df1 menggunakan fuzzy matching pada kolom Nama.
    threshold: skor kecocokan minimum (0-100), makin tinggi makin ketat
    """

    # Kolom dari df1 yang ingin ditambahkan ke df2
    cols_to_add = ['Jenis Kelamin', 'Formasi', 'Penempatan']

    matched_rows = []

    for nama in df2[col]:
        # Cari nama paling mirip di df1
        result = process.extractOne(
            nama,
            df1[col].tolist(),
            scorer=fuzz.token_sort_ratio  # Bagus untuk nama yang urutannya berbeda
        )

        if result and result[1] >= threshold:
            matched_nama = result[0]
            score = result[1]
            row = df1[df1[col] == matched_nama][cols_to_add].iloc[0].to_dict()
            row['_matched_nama'] = matched_nama  # Untuk verifikasi
            row['_match_score'] = score
        else:
            # Tidak ditemukan kecocokan
            row = {c: None for c in cols_to_add}
            row['_matched_nama'] = None
            row['_match_score'] = None

        matched_rows.append(row)

    result_df = pd.DataFrame(matched_rows)
    df2 = pd.concat([df2.reset_index(drop=True), result_df], axis=1)
    return df2


# Jalankan
df2_merged = fuzzy_merge(df2, df1, threshold=80)

# Cek hasil — terutama yang score-nya tidak 100 (perlu diverifikasi manual)
print(df2_merged[['Nama', '_matched_nama', '_match_score']].sort_values('_match_score'))
```
