```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_excel("/content/Kelas Disiplin Feb 2025.xlsx")
df2 = pd.read_excel("/content/Piket Kelas Disiplin Feb 2025.xlsx")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1['respon_menyalami_guru'].fillna(df1['respon_menyalami_guru'].mode()[0], inplace=True)
```

```python
def hapus_setelah_titik(str):
    return str.split(".")[0]
```

```python
df1.info()
```

```python
df2.info()
```

```python
for m in range (3, 13):
  df1[df1.columns[m]] = df1[df1.columns[m]].apply(hapus_setelah_titik)
```

```python
for n in range (3, 9):
  df2[df2.columns[n]] = df2[df2.columns[n]].apply(hapus_setelah_titik)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df1.head()
```

```python
df2.head(5)
```

```python
df1["keaktifan"].replace({"1": "25", "2": "50", "3": "75", "4": "100"}, inplace=True)
```

```python
kolom_doa = ['guru_doa', 'murid_doa']

for x in kolom_doa:
  df1[x].replace({"1": "50", "2": "100"}, inplace=True)
```

```python
kolom1 = ['izin_kbm',
       'siswa_masuk_kelas', 'respon_papan_tulis',
       'respon_meja_kursi', 'respon_membangunkan', 'respon_menyalami_guru',
       'kelengkapan_sepatu']

for x in kolom1:
  df1[x].replace({"1": "33.333", "2": "66.666", "3": "99.999"}, inplace=True)
```

```python
kolom2 = ['kebersihan', 'respon_penanganan_sampah',
       'kelengkapan_inventaris', 'kerapian', 'kelengkapan_seragam',
       'kerapian_akhir']

for x in kolom2:
  df2[x].replace({"1": "33.333", "2": "66.666", "3": "99.999"}, inplace=True)
```

```python
for y in range (3, 13):
  df1[df1.columns[y]] = df1[df1.columns[y]].astype(float)

for y in range (3, 9):
  df2[df2.columns[y]] = df2[df2.columns[y]].astype(float)
```

```python
df1.drop(columns=(['tanggal', 'guru']), inplace=True)
df2.drop(columns=(['tanggal', 'guru']), inplace=True)
```

```python
agg1 = df1.groupby(['kelas']).mean()
agg1.head(10)
```

```python
agg2 = df2.groupby(['kelas']).mean()
agg2.head(10)
```

```python
agg2_mean = agg2.mean()
agg2_mean
```

```python
agg1.index
```

```python
#Karena tidak ada data kebersihan untuk kelas TQM, maka data akan diisi dengan memasukkan rata2 keseluruhan

new_TQM = pd.DataFrame([agg2_mean], index=['10 (TQM)'])
new_TQM
```

```python
#Menambahkan hasil rata-rata ke dalam dataframe

agg2_new = pd.concat([new_TQM, agg2])
agg2_new
```

```python
#gabungkan kolom untuk kelas/baris yang sama

df = pd.concat([agg1, agg2_new], axis=1)
df
```

```python
#kelompokkan setiap kategori ke dalam kategori utama
df["kebersihan_total"] = df["kebersihan"]
df["keaktifan_total"] = (df["keaktifan"] + df["izin_kbm"] + df["siswa_masuk_kelas"])/3
df["berdoa"] = (df["guru_doa"] + df["murid_doa"])/2
df["responsif_total"] = (df["respon_papan_tulis"] + df["respon_meja_kursi"] + df["respon_membangunkan"] + df["respon_menyalami_guru"] + df["respon_penanganan_sampah"])/5
df["kerapian_total"] = (df["kerapian"] + df["kerapian_akhir"])/2
df["kelengkapan_total"] = (df["kelengkapan_sepatu"] + df["kelengkapan_inventaris"] + df["kelengkapan_seragam"])/3
```

```python
#buat dataframe baru berdasarkan kategori utama
df_final = df[['kebersihan_total',
       'keaktifan_total', 'berdoa', 'responsif_total', 'kerapian_total',
       'kelengkapan_total']]

df_final.head(10)
```

```python
#menambahkan kolom total berdasarkan rata2 keseluruhan aspek/kategori
df_final["total_avg"] = df_final.mean(axis=1)
df_final.head(10)
```

```python
#sortir berdasarkan total tertinggi
df_final.sort_values(by="total_avg", ascending=False).round(3)
```

```python
sns.barplot(data=df_final['total_avg'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin Februari 2025 [Keseluruhan]')
plt.xlabel('Rata-Rata Akhir')
plt.ylabel('Kelas');
```

```python
sns.barplot(data=df_final['keaktifan_total'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin Februari 2025 [Keaktifan]')
plt.xlabel('Rata-Rata Akhir')
plt.ylabel('Kelas');
```

```python
sns.barplot(data=df_final['kebersihan_total'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin Februari 2025 [Kebersihan]')
plt.xlabel('Rata-Rata Akhir')
plt.ylabel('Kelas');
```
