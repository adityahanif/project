```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_excel("/content/disiplin kelas november.xlsx")
df2 = pd.read_excel("/content/piket disiplin kelas november.xlsx")
```

```python
df1['respon_menyalami_guru'].fillna(df1['respon_menyalami_guru'].mode()[0], inplace=True)
```

```python
def hapus_setelah_titik(str):
    return str.split(".")[0]
```

```python
for y in range (3, 13):
  df1[df1.columns[y]] = df1[df1.columns[y]].apply(hapus_setelah_titik)
```

```python
for y in range (3, 9):
  df2[df2.columns[y]] = df2[df2.columns[y]].apply(hapus_setelah_titik)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df1["keaktifan"].replace({"1": "25", "2": "50", "3": "75", "4": "100"}, inplace=True)
```

```python
kolom_doa = ['guru_doa', 'murid_doa']

for x in kolom_doa:
  df1[x].replace({"1": "50", "2": "100"}, inplace=True)
```

```python
kolom1 = ['izin_kbm',
       'siswa_masuk_kelas', 'respon_papan_tulis',
       'respon_meja_kursi', 'respon_membangunkan', 'respon_menyalami_guru',
       'kelengkapan_sepatu']

for x in kolom1:
  df1[x].replace({"1": "33.333", "2": "66.666", "3": "99.999"}, inplace=True)
```

```python
kolom2 = ['kebersihan', 'respon_penanganan_sampah',
       'kelengkapan_inventaris', 'kerapian', 'kelengkapan_seragam',
       'kerapian_akhir']

for x in kolom2:
  df2[x].replace({"1": "33.333", "2": "66.666", "3": "99.999"}, inplace=True)
```

```python
for y in range (3, 13):
  df1[df1.columns[y]] = df1[df1.columns[y]].astype(float)

for y in range (3, 9):
  df2[df2.columns[y]] = df2[df2.columns[y]].astype(float)
```

```python
df2.drop(columns=(['tanggal', 'guru']), inplace=True)
```

```python
agg1 = df1.groupby(['kelas']).mean()
agg1
```

```python
agg2 = df2.groupby(['kelas']).mean()
agg2
```

```python
df = pd.concat([agg1, agg2], axis=1)
df
```

```python
df.shape
```

```python
df["kebersihan_total"] = df["kebersihan"]
df["keaktifan_total"] = (df["keaktifan"] + df["izin_kbm"] + df["siswa_masuk_kelas"])/3
df["berdoa"] = (df["guru_doa"] + df["murid_doa"])/2
df["responsiif_total"] = (df["respon_papan_tulis"] + df["respon_meja_kursi"] + df["respon_membangunkan"] + df["respon_menyalami_guru"] + df["respon_penanganan_sampah"])/5
df["kerapian_total"] = (df["kerapian"] + df["kerapian_akhir"])/2
df["kelengkapan_total"] = (df["kelengkapan_sepatu"] + df["kelengkapan_inventaris"] + df["kelengkapan_seragam"])/3
```

```python
df.columns
```

```python
df_final = df[['kebersihan_total',
       'keaktifan_total', 'berdoa', 'responsiif_total', 'kerapian_total',
       'kelengkapan_total']]

df_final
```

```python
df_final["total_avg"] = df_final.mean(axis=1)
df_final["total_sum"] = df_final.sum(axis=1)
```

```python
df_final.sort_values(by="total_avg", ascending=False).round(3)
```

```python
sns.barplot(data=df_final['total_avg'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin November [Keseluruhan]')
plt.xlabel('Rata-Rata Akhir')
plt.ylabel('Kelas');
```

```python
sns.barplot(data=df_final['keaktifan_total'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin November [Keaktifan]')
plt.xlabel('Rata-Rata Akhir')
plt.ylabel('Kelas');
```

```python
sns.barplot(data=df_final['kebersihan_total'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin November [Kebersihan]')
plt.xlabel('Rata-Rata Akhir')
plt.ylabel('Kelas');
```

```python
sns.barplot(data=df_final['kerapian_total'].sort_values(ascending=False), orient='h')
plt.title('Grafik Lomba Kelas Disiplin November [Kerapian]')
plt.xlabel('Rata-Rata Akhir')
plt.ylabel('Kelas');
```
