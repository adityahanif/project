```python
import pdfplumber
import pandas as pd

pdf_path = "/content/integrasi.pdf"

filtered_tables = []

with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        table = page.extract_table()

        if table:
            df = pd.DataFrame(table[1:], columns=table[0])

            # Normalize column count (handle None headers)
            num_cols = len(df.columns)

            if num_cols == 7:
                print(f"Page {i+1}: ✅ Lanjut (7 columns)")
                filtered_tables.append(df)
            else:
                print(f"Page {i+1}: ❌ Skip ah ({num_cols} columns)")

# Merge all valid tables
if filtered_tables:
    merged_df = pd.concat(filtered_tables, ignore_index=True)
else:
    merged_df = pd.DataFrame()

print("Final shape:", merged_df.shape)

merged_df.to_excel("merged_df_original_file.xlsx", index=False)
```

```python
df_mentah.rename(columns={
    'Tanggal\nLahir': "Tanggal Lahir",
    'Nilai/IPK': 'IPK',
    'Nilai SKD': 'TWK',
    'Unnamed: 7': 'TIU',
    'Unnamed: 8': 'TKP',
    'Unnamed: 9': 'SKD',
    'Skor SKD\n(40%)': 'SKD 40%',
    'Nilai\nSKB': 'SKB',
    'Skor SKB\n(60%)': 'SKB 60%',
    'Nilai\nAkhir': 'Nilai Akhir'}, inplace=True)

df_mentah.head(10)
```

```python
df_mentah = df_mentah.dropna()

df_mentah.head(10)
```

```python
df_mentah[df_mentah['Nama'].str.contains('3', na=False)]
```

```python
df_mentah = df_mentah[~df_mentah['Nama'].str.contains('3', na=False)]

df_mentah.head(10)
```

```python
df_mentah = df_mentah.replace(r'\n', ' ', regex=True)

df_mentah.head(10)
```

```python
df_mentah.drop(columns=['No'], inplace=True)
df_mentah.head(10)
```

```python
cols_numeric = [
    'IPK', 'TWK', 'TIU', 'TKP',
    'SKD', 'SKD 40%', 'SKB',
    'SKB 60%', 'Nilai Akhir'
]

df_mentah[cols_numeric] = df_mentah[cols_numeric].apply(
    pd.to_numeric, errors='coerce'
)

df_mentah.info()
```
